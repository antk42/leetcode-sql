```python
sql = '''
SELECT * 
FROM

(SELECT e.*, d.name,
    dense_rank() OVER (PARTITION BY e.departmentId ORDER BY e.salary DESC) as rnk
FROM Employee AS e
LEFT JOIN Department AS d ON e.departmentId = d.id) AS t
WHERE t.rnk < 3
'''
```

```python
select(sql)
```

```python
sql = '''
SELECT * 
FROM

(SELECT e.*, d.name as department,
    dense_rank() OVER (PARTITION BY e.departmentId ORDER BY e.salary DESC) as rnk
FROM Employee AS e
LEFT JOIN Department AS d ON e.departmentId = d.id) AS t
WHERE t.rnk < 3
'''
```

```python
select(sql)
```
